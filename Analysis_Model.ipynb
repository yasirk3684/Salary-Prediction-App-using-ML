##  Model Training Script From Here

```python
df = df[[
    "Years",
    "Job Rate",
    "Overtime Hours",
    "Department",
    "Country",
    "Monthly Salary"
]]
```

###  Feature / Target Split

```python
X = df.drop("Monthly Salary", axis=1)
y = df["Monthly Salary"]
```

###  Identify Column Types

```python
categorical_cols = ["Department", "Country"]
numeric_cols = ["Years", "Job Rate", "Overtime Hours"]
```

###  Preprocessing

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.compose import ColumnTransformer
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_cols),
        ("num", "passthrough", numeric_cols)
    ]
)
```

###  Build ML Pipeline

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("regressor", LinearRegression())
])
```

###  Train Model 

```python
model.fit(X, y)
```

###  Save Model 

```python
import joblib
```

```python
joblib.dump(model, "salary_model_final.pkl")

print("âœ¨ Training complete! Model saved as salary_model_final.pkl")
```
